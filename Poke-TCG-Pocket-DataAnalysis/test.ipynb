```python
import pandas as pd
```

```python
df = pd.read_csv('simulation/data/pokemon-tcg-with-fossils.csv')
df.head(5)
```

```python
df['ability'].unique()
```

```python
df[df['name'].str.contains('darkrai')]
```

```python
abilities = df[~df['ability'].isna()][['id', 'name', 'ability']]
abilities = abilities.reset_index().drop(columns=['index'])
abilities
```

```python
output = "id_abilities = {\n"
for i in range(abilities['id'].count()):
    id = abilities.loc[i, 'id']
    name = abilities.loc[i, 'ability'].replace(' ', '_')
    output += f"    '{id}': {name},\n"

output += "}"
print(output)
```
